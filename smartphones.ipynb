```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
path=r"C:\Users\Mini\Downloads\chromedriver_win32\chromedriver.exe"
```

```python
driver = webdriver.Chrome(path)
```

```python
driver.get('https://www.amazon.in/s?k=mobile+phones+under+20000&rh=n%3A1389432031&dc&qid=1604132579&rnid=3576079031&ref=sr_nr_n_1')
```

```python
content=driver.page_source
soup= BeautifulSoup(content)
names=soup.findAll('span',attrs={'class':'a-size-medium a-color-base a-text-normal'})
print(len(names))
amount=soup.findAll('span',attrs={'class':'a-price-whole'})
print(len(amount))
image=soup.findAll('img',attrs={'s-image'})
print(len(image))
rating=soup.findAll('span',attrs={'class':'a-icon-alt'})
print(len(rating))
```

```python
Products=[]
Price=[]
Image=[]
Rating=[]
```

```python
names
```

```python
amount
```

```python
image
```

```python
rating
```

```python
for i in names:
    Products.append(i.text)
Products
```

```python
df=pd.DataFrame(Products)
df=df.rename({0:'Product Name'},axis=1)
df
```

```python
for i in amount:
    Price.append(i.text)
Price
```

```python
df1=pd.DataFrame(Price)
df1=df1.rename({0:'Price'},axis=1)
df1
```

```python
for i in image:
    Image.append(i.get('src'))
print(len(Image))
Image
```

```python
df2=pd.DataFrame(Image)
df2=df2.rename({0:'Image Url'},axis=1)
df2=df2.iloc[0:27]
df2
```

```python
for i in rating:
    Rating.append(i.text)
Rating
```

```python
df3=pd.DataFrame(Rating)
df3=df3.rename({0:'Rating'},axis=1)
df3=df3.iloc[4:]
df3=df3.reset_index(drop=True)
df3
```

```python
Smartphone=pd.concat([df,df1,df2,df3],axis=1)
Smartphone.index=range(24)
Smartphone
```

```python
Smartphone.to_csv("Smartphones_under_20000.csv")
```
